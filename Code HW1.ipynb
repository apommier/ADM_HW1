```python
# Text Wrap
import textwrap

def wrap(string, max_width):
    N=len(string)
    res=''
    for i in range(0,N+1,max_width):
        if i == 0 :
            res=res+string[i:i+max_width]
        else :
            res=res+'\n'+string[i:i+max_width]
    return res

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```

```python
# String Formatting

def print_formatted(number):
    max_width = len(bin(number)[2:])
    for i in range(1, number+1):
        decimal = str(i)
        octal = oct(i)[2:]
        hexadecimal = hex(i)[2:].upper() #to have uppercase
        binary = bin(i)[2:]
        print(decimal.rjust(max_width),octal.rjust(max_width),hexadecimal.rjust(max_width),binary.rjust(max_width))

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

```python
# Alphabet Rangoli

def print_rangoli(size):
    for i in range(size):
        l=[]
        for j in range(i+1):
            l.append(chr(ord('a')+size-j-1))
        s = "-".join(l)
        print((s+s[::-1][1:]).center(size*4-3, '-'))

    for i in range(size-1):
        l=[]
        for j in range(size-i-1):
            l.append(chr(ord('a')+size-j-1))
        s = "-".join(l)
        print((s+s[::-1][1:]).center(size*4-3, '-'))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
# the Minion game
def minion_game(string):
    stuart=0
    kevin=0
    vowels=['A','E','I','O','U']
    for i in range(len(string)):
        score=len(string)-i
        if string[i] in vowels:
            kevin+=score
        else : 
            stuart+=score
    if kevin == stuart :
        print("Draw")
    elif kevin > stuart :
        print("Kevin",kevin)
    else :
        print("Stuart", stuart)

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
## Collections.OrderedDict()
from collections import OrderedDict
OD = OrderedDict()
N=int(input())
for i in range(N):
    # info about the item
    product=input().split()
    price = int(product[-1])
    # join every words of the name
    name =" ".join(product[:-1])
    # see if this item is already in OD
    if OD.get(name) :
        # add the amount of the price
        OD[name] += price
    else :
        # create the item in OD and initialize the price
        OD[name] = price
for name in OD.keys():
    print(name, OD[name])
```

```python
## Piling Up!

from collections import deque
T=int(input())

# for each cube
for i in range(T):
    n=int(input())
    d = deque(map(int,input().split()))
    while d:
        # test if we have to choose the right or the left (the biggest)
        if d[0]>d[-1]:
            big=d.popleft()  
        else : 
            big=d.pop()
            
        # test if there still is a number in the deque
        if len(d)==0:
            print("Yes")
            break
        # test if I can continue the pile
        elif d[-1]>big or d[0]>big:
            print("No")
            break
```

```python
## Time Delta

from datetime import datetime

def time_delta(t1, t2):
    # read the right values of the string in fixing the format (documentation of Datetime)
    t1 = datetime.strptime(t1, '%a %d %b %Y %H:%M:%S %z')
    t2 = datetime.strptime(t2, '%a %d %b %Y %H:%M:%S %z')
    # count the second between the two dates
    res=(t1-t2).total_seconds()
    return str(int(abs(res)))

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    t = int(input())
    for t_itr in range(t):
        t1 = input()
        t2 = input()
        delta = time_delta(t1, t2)
        fptr.write(delta + '\n')
    fptr.close()
```

```python
## XML 1 - Find the Score
import sys
import xml.etree.ElementTree as etree

def get_attr_number(node):
    # i start with the number of attributes in the current node (the greatest) stored in a dictionary
    res = len(node.attrib)
    # it is a tree, so the main node relate/contain the others 
    for ele in node:
        # i use recursivity to get the number of attributes pf the child node
        res += get_attr_number(ele) 
    return res

if __name__ == '__main__':
    sys.stdin.readline()
    xml = sys.stdin.read()
    tree = etree.ElementTree(etree.fromstring(xml))
    root = tree.getroot()
```

```python
## XML2 - Find the Maximum Depth
import xml.etree.ElementTree as etree

maxdepth = 0
def depth(elem, level):
    global maxdepth
    #if an e as the same level as maxdepth i don't add 1
    if level == maxdepth:
        maxdepth+=1
    for e in elem :
        # i test the child node of elem to know if they have the same level and else I increase the maxdepth
        depth(e, level+1)

if __name__ == '__main__':
    n = int(input())
    xml = ""
    for i in range(n):
        xml =  xml + input() + "\n"
    tree = etree.ElementTree(etree.fromstring(xml))
    depth(tree.getroot(), -1)
    print(maxdepth)
```

```python
## Standardize Mobile Number Using Decorators

def wrapper(f):
    def fun(l):
        # i keep only the 10 last digits
        number_without_prefix=[number[-10:] for number in l]
        # i change the format
        list_numbers= ['+91'+' '+number[0:5]+' '+number[5:] for number in number_without_prefix]
        f(list_numbers)
    return fun

@wrapper
def sort_phone(l):
    print(*sorted(l), sep='\n')

if __name__ == '__main__':
    l = [input() for _ in range(int(input()))]
    sort_phone(l) 
```

```python
## Number Line Jumps
import math
import os
import random
import re
import sys

def kangaroo(x1, v1, x2, v2):
    #if x1 is behind x2
    if x1<x2 :
        if v1<=v2 : 
            # x1 will not be able to catch up with x2 (it is behind and slower or equal speed)
            return 'NO'
        # while x1 don't overtake x2 we continue
        while x1<x2:
            # increase a jump
            x1=x1+v1
            x2=x2+v2
            if x1==x2:
                # return YES if they are in the same location after the jump
                return 'YES'
    # if they start at the same location, it is good
    elif x1==x2:
        return 'YES'
    else:
        if v1<=v2 : 
            # x2 will not be able to catch up with x1 (it is behind and slower or equal speed)
            return 'NO'
        while x2<x1:
            x1=x1+v1
            x2=x2+v2
            if x1==x2:
                # return YES if they are in the same location after the jump
                return 'YES'
    # else they can't be at the same location
    return 'NO'

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    first_multiple_input = input().rstrip().split()
    x1 = int(first_multiple_input[0])
    v1 = int(first_multiple_input[1])
    x2 = int(first_multiple_input[2])
    v2 = int(first_multiple_input[3])
    result = kangaroo(x1, v1, x2, v2)
    fptr.write(result + '\n')
    fptr.close()
```
